```python
import pandas as pd
import re  
import string  

# Load the dataset
df = pd.read_csv('merged_files.csv')

# Define the cleaning function
def clean_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove emojis and special unicode characters
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  
                               u"\U0001F300-\U0001F5FF"  
                               u"\U0001F680-\U0001F6FF"  
                               u"\U0001F1E0-\U0001F1FF"  
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Strip whitespace
    text = text.strip()
    return text

# Concatenate all text columns into one column and handle NaN values
df['combined_comment'] = df.fillna('').agg(lambda x: ' '.join(x.values.astype(str)), axis=1)

# Clean the combined text
df['combined_comment'] = df['combined_comment'].apply(clean_text)


df['data'] = df['combined_comment'].apply(lambda x: f"comment: {x}")


df_final = df[['data']].copy()

# Save to a new CSV file
df_final.to_csv('merged_files22.csv', index=False)


df_final.head()



```

```python
def clean_and_tokenize(data):
    # Lowercase the text
    data = data.lower()
    # Remove the word 'comment'
    data = data.replace('comment', '')
    # Tokenize the text
    tokens = word_tokenize(data)
    # Prepare a translation table to remove punctuation
    table = str.maketrans('', '', string.punctuation)
    # Remove stop words and punctuation
    stop_words_set = set(stopwords.words('english'))
    tokens = [word.translate(table) for word in tokens if word.isalpha() and word not in stop_words_set]
    # Return the list of filtered tokens
    return tokens

# Apply the function to each row in the 'data' column of the dataframe
df_final['tokens'] = df_final['data'].apply(clean_and_tokenize)

# Explode the 'tokens' lists into separate rows
df_exploded = df_final.explode('tokens')

# Save the tokenized data to a new CSV file, with each token on a separate row
df_exploded.to_csv('tokenized_data.csv', columns=['tokens'], index=False)

# Display the first few rows to verify the changes
df_exploded.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction import text 

# Load the CSV file
file_path = 'tokenized_data.csv'  
data = pd.read_csv(file_path)

# Remove any NaN values in the data
data_cleaned = data.dropna()

# Define stop words. Adding custom stop words to the default English stop words list
additional_stop_words = ['like', 'just', 'said', 'think', 'know', 'people', 'time', 'year']
stop_words = list(text.ENGLISH_STOP_WORDS.union(additional_stop_words))

# Initializing the TfidfVectorizer with the custom stop words
vectorizer = TfidfVectorizer(stop_words=stop_words, use_idf=True)

# Applying TF-IDF transformation to the cleaned data
tfidf_matrix = vectorizer.fit_transform(data_cleaned['tokens'])

# Extract words and their corresponding TF-IDF scores
words = vectorizer.get_feature_names_out()
scores = tfidf_matrix.sum(axis=0).A1
tfidf_scores = pd.DataFrame(list(zip(words, scores)), columns=['Word', 'TF-IDF']).sort_values(by='TF-IDF', ascending=False)


# Display the top meaningful words
print(tfidf_scores.head(30))


```

```python
from textblob import TextBlob
import pandas as pd

# Example words from your dataset
words = ['cancer', 'god', 'love', 'prayers', 'hope', 'beautiful', 'pray', 'cure']

# Create a DataFrame to store the results
sentiments = pd.DataFrame(columns=['Word', 'Polarity', 'Subjectivity'])

# Analyze the sentiment of each word
for word in words:
    blob = TextBlob(word)
    sentiment = blob.sentiment
    # Create a new DataFrame for the current word and its sentiments
    new_row = pd.DataFrame({'Word': [word], 'Polarity': [sentiment.polarity], 'Subjectivity': [sentiment.subjectivity]})
    # Concatenate the new DataFrame to the existing one
    sentiments = pd.concat([sentiments, new_row], ignore_index=True)

print(sentiments)

```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Example words
words = ['cancer', 'god', 'love', 'prayers', 'hope', 'beautiful', 'pray', 'cure']

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Create a DataFrame to store the results
sentiments_vader = pd.DataFrame(columns=['Word', 'Negativity', 'Neutrality', 'Positivity', 'Compound'])

# Analyze the sentiment of each word
for word in words:
    vs = analyzer.polarity_scores(word)
    new_row = pd.DataFrame({
        'Word': [word],
        'Negativity': [vs['neg']],
        'Neutrality': [vs['neu']],
        'Positivity': [vs['pos']],
        'Compound': [vs['compound']]
    })
    # Concatenate the new row to the existing DataFrame
    sentiments_vader = pd.concat([sentiments_vader, new_row], ignore_index=True)

print(sentiments_vader)


```

```python
import csv
import re
import nltk
from nltk.corpus import stopwords
from collections import Counter

# Download NLTK stopwords data
nltk.download('stopwords')

# Read the text file
with open('tokens.txt', 'r', encoding='utf-8') as file:
    text = file.read()


# Tokenize the text into words
words = nltk.word_tokenize(text)


# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]

# Count the occurrences of each meaningful word
word_counts = Counter(filtered_words)

# Write word counts to a CSV file
with open('meaningful_word_counts.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Word', 'Count'])
    for word, count in word_counts.items():
        writer.writerow([word, count])

```
